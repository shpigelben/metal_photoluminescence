```python
import matplotlib.pyplot as plt
import numpy as np

def gaussian(x, mu, sigma):
    E = (x-mu)/(np.sqrt(2)*sigma)
    return 1/(np.sqrt(2*np.pi)*sigma) *  np.exp(-E**2)

x = np.linspace(-10, 10, 100)
mu = 0
sigma = 1

y = gaussian(x, mu, sigma)


plt.plot(x, y)
plt.show()
```

```python
def integrand(x, mu, sigma):
    return x * gaussian(x, mu, sigma)

mu = 3
sigma = 0.1
a = mu - 5*sigma
b = mu + 5*sigma
N = 100
x = np.linspace(a, b, N)


for sigma in [0.3, 0.2, 0.1]:
    y = integrand(x, mu, sigma)
    I = np.trapezoid(y, x)
    print(I)
```
